```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
x = np.arange(0.001, 4, 0.1)
#y = np.log(15*x) -2*x**2

def fun(a):
    return np.log(15*a) -2*a**2

plt.figure()
plt.plot(x, fun(x), linewidth = 1)
plt.scatter(0.5, fun(0.5))
plt.xlabel('x')
plt.ylabel('log$(15x) - 2x^2$')
```

```python
fig = plt.figure()

ax = Axes3D(fig)

l1 = 0
u1 = 3
x1 = np.arange(l1, u1, 0.25)
x2 = np.arange(l1, u1, 0.25)
X1, X2 = np.meshgrid(x1, x2)
#plt.fill_between(x1, 5-0.5*x1,  alpha = 0.3, color = 'red')
y = 2*np.ones(len(x1))

def fun(x, y):
    return  - (x-2)**2 - (y -2)**2 +10 
V = fun(X1, X2)
ax.plot_wireframe(X1, X2, V,  rstride = 1, cstride = 1, linewidth = 0.5)
#

plt.plot(x1, y, ':',  c = 'red')
plt.plot(x1, y, 10*np.ones(len(x1)), ':',  c = 'red')
plt.plot(x1, y, fun(x1, y), '-',  c = 'red')

plt.plot(y, x1,':',  c = 'green')
plt.plot(y, x1,10*np.ones(len(x1)), ':',  c = 'green')
#plt.plot([2], [2], 'o', c = 'red')

plt.plot(y, x1, fun(y, x1), '-',  c = 'green')
plt.plot([2], [2] , fun(2, 2), 'o', c = 'red')


plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
ax.zaxis.set_rotate_label(False) 
ax.set_zlabel('$f({x_1}, x_2)$', rotation = 90)
ax.view_init(30, 260)
#plt.contour(X1, X2, V, cmap = 'viridis')
#plt.contour(X1, X2, V, 50,  linewidths = 0.5, colors = 'k')
name = 'Multivariate FOC'
plt.title(name)
#fig.savefig(name + '.png')
```
